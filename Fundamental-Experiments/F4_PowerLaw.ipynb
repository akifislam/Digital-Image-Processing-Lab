```python
import cv2
import matplotlib.pyplot as plt 
import numpy as np

img = cv2.imread("PowerLaw_TestImage.png")
img = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

plt.imshow(img,cmap='gray',vmin=0,vmax=255)
plt.show()
```

# Power Law

```python
img = img/255.0 

img = img**(5)

plt.imshow(img,cmap='gray')
plt.show()
```
