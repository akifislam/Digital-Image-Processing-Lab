```python
import cv2
img = cv2.imread("LUNA_NORMAL.jpeg")
img = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
# img = cv2.resize(img,(512,512))

import matplotlib.pyplot as plt 
plt.imshow(img,cmap='gray')
plt.show()
```

# Negative Image Algorithm

```python
import numpy as np
```

```python
max_intensity = np.max(img)
print(max_intensity)
```

```python
height,width = img.shape
```

```python
def make_negative(img):

    for i in range(0,height):
        for j in range(0,width):
            img[i][j] = max_intensity - img[i][j]
            
    return img
```

# 1st time Negative

```python
import matplotlib.pyplot as plt 
plt.imshow(make_negative(img),cmap='gray')
plt.show()
```

# 2nd Time Negative

```python
import matplotlib.pyplot as plt 
plt.imshow(make_negative(img),cmap='gray')
plt.show()
```
