# Load a White Circle of 400x400

```python
import cv2
img = cv2.imread("white_circle_big.png")

import matplotlib.pyplot as plt
plt.imshow(img)
plt.show()
```

```python
print(img.shape)
```

# Load the same Circle of 10x10

```python
img = cv2.imread("white_circle_small.png")

plt.imshow(img)
plt.show()
```

```python
img = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
img.shape
```

# Matrix Representation of a 10x10 Image

```python
print(img)
```

# Image Manipulation in the Matrix

```python
img[3] #4th Row
```

# Filling all values of 4th Row with intensity 255

```python
for i in range(0,len(img[3])):
   if img[3][i] == 0:
       img[3][i] = 255
```

```python
img[3]
```

# Plot

```python
plt.imshow(img,cmap='gray')
plt.show()
```
