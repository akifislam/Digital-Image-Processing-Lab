# Manual Histogram Function

```python
import matplotlib.pyplot as plt 

def show_histogram(img):

    occurance = {}
    
    for row in img:
        for value in row:
                
            value = np.floor(value)
            
            if value in occurance:
                occurance[value]+=1
            else:
                occurance[value]=1

    sorted_occurance = {k: occurance[k] for k in sorted(occurance)}
    x_values = sorted_occurance.keys()
    y_values = sorted_occurance.values()

    plt.bar(x_values,y_values)


    # Built-in Function
    # plt.hist(img.ravel(),bins=256,range=(0,256))
    # plt.show()
```

# Load Image of Lena and Show Histogram 

```python
import cv2
import matplotlib.pyplot as plt 
import numpy as np
```

```python
def show_image(img):
    plt.imshow(img,cmap='gray')
```

```python
color_img = cv2.imread("Lec3_Lena.png")
gray_img = cv2.cvtColor(color_img,cv2.COLOR_BGR2GRAY)


plt.figure(figsize=(15,4))

plt.subplot(1,2,1)
show_image(gray_img)

plt.subplot(1,2,2)

show_histogram(gray_img)


plt.title("Lena Histogram")

plt.show()

```

# Negative Image

```python
def make_negative(imgg):

    h,w = imgg.shape
    for i in range(0,h):
        for j in range(0,w):
            imgg[i][j] = 1 - imgg[i][j]
            
    return imgg
```

```python
original_img = img
print(np.max(original_img))
```

```python
print(img[0][0])
```

```python
neg_img = make_negative(original_img)
print(neg_img[0][0])
```

# Histogram Equilization using Built in Function

```python
eq_hist_img = cv2.equalizeHist(original_img)
eq_hist_img = make_negative(eq_hist_img)

plt.imshow(eq_hist_img,cmap='gray',vmin=0,vmax=256)
plt.show()

plt.hist(eq_hist_img.flatten(),256,[0,256])
plt.show()
```
