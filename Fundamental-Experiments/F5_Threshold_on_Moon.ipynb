```python
import cv2
import matplotlib.pyplot as plt 
import numpy as np

img = cv2.imread("Thresholding_Test.png")
img = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

plt.imshow(img,cmap='gray')
plt.show()
```

# Thresholding

```python
img[img>100] = 255
img[img<=125] = 0

plt.imshow(img,cmap='gray')
plt.show()
```
