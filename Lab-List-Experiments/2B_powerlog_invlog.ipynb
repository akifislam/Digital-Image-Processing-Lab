# Power Law Transform

```python
import cv2
img = cv2.imread("../mycat.jpg")
# img = cv2.imread("white_circle_small.png")

img = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
img = cv2.resize(img,(512,512))
```

```python
import matplotlib.pyplot as plt 
plt.imshow(img,cmap="gray")
plt.show()
```

```python
height,width = img.shape
```

```python
img
```

```python
# Power Law
import numpy as np 

def apply_powerlaw_transform(imgg,factor):
    
    imgg = imgg/255
    
    for i in range(0,height):
        for j in range(0,width):
            imgg[i][j] = (imgg[i][j]**factor)

    imgg = imgg/255

    return imgg
```

```python
plaw_img = apply_powerlaw_transform(img,1.55)
plt.imshow(plaw_img,cmap='gray')
plt.show()
```

# Inverse Log Transform

```python
def apply_inverse_log_transform(imgg, factor):

    imgg = imgg/255
    
    for i in range(0,height):
        for j in range(0,width):
            imgg[i][j] = (np.exp(imgg[i][j])**(1/factor))-1

    imgg = imgg*255

    return imgg
        
```

```python
plt.imshow(apply_inverse_log_transform(plaw_img,5),cmap='gray')
plt.show()
```
