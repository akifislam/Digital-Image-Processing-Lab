# Load Image

```python
import cv2
img = cv2.imread("Mini.jpg")
img = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
img = cv2.resize(img,(512,512))
```

# Original Image Plot 512x512

```python
import matplotlib.pyplot as plt 
plt.imshow(img,cmap="gray")
plt.show()
```

```python
height,width = img.shape
```

# Downsampling

```python
import numpy as np

def downsampling(decrease_by):
    new_image = []

    for i in range(0,height,decrease_by):
        temp_row = []
        for j in range(0,width,decrease_by):       
            temp_row.append(img[i][j])
        
        new_image.append(temp_row)

    new_image = np.array(new_image)    
    return new_image
```

# Subplots

```python
plt.figure(figsize=(10,6))

for i in range(1,9):
    plt.subplot(2,4,i)
    x = downsampling(2**(i-1)) #(i-1)=0 provides the original image
    h,w = x.shape
    plt.imshow(x,cmap='gray')
    plt.title(f"{h}x{w}")

plt.show()
```
