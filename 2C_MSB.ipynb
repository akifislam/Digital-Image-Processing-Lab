```python
import cv2
img = cv2.imread("../mycat.jpg")
img = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
img = cv2.resize(img,(512,512))
```

```python
import matplotlib.pyplot as plt 
plt.imshow(img,cmap="gray")
plt.show()
```

# MSB

```python
import numpy as np

height,width = img.shape

def MSB(imgg):
    mask = 0b11100000
    imgg = np.bitwise_and(imgg, mask)
    return imgg
```

```python
MSB_img = MSB(img)
plt.imshow(MSB_img,cmap='gray')
plt.show()
```

```python
dif_img = img-MSB_img
plt.imshow(dif_img,cmap='gray')
plt.show()
```
