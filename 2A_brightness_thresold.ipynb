# Load Image

```python
import cv2
img = cv2.imread("Mini.jpg")
# img = cv2.imread("white_circle_small.png")

img = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
img = cv2.resize(img,(512,512))
```

```python
img
```

```python
height,width = img.shape
img.shape
```

# Brightness Change Function

```python
import numpy as np 

def change_brightness(imgg,x):
    
    imgg = imgg.astype(np.uint16)
    # imgg += x
    # imgg[imgg > 255] = 255
    for i in range(0,height):

        for j in range(0,width):
         
            imgg[i][j]+=x
 
            if imgg[i][j] > 255:
                imgg[i][j] = 255

            if imgg[i][j] < 0:
                imgg[i][j] = 0

    return imgg.astype(np.uint8)
```

# Original Image

```python
import matplotlib.pyplot as plt
plt.imshow(img,cmap='gray')
plt.show()
```

```python
img
```

### Change Matrix

```python
new_img = change_brightness(img,100)
new_img
```

```python
plt.imshow(new_img,cmap='gray')
plt.show()
```

# Subplot (Positive Brightness)

```python
plt.figure(figsize=(9,12))
cnt = 0

for i in range(1,10):
    plt.subplot(3,3,i)
    plt.imshow(change_brightness(img,cnt),cmap="gray")
    plt.title(f"+{cnt}")
    cnt+=20
    
plt.show()
```

# Subplot (Negative Brightness)

```python
plt.figure(figsize=(9,12))
cnt = 0

for i in range(1,10):
    plt.subplot(3,3,i)
    plt.imshow(change_brightness(img,cnt),cmap="gray")
    plt.title(f"-{cnt}")
    cnt-=20
    
plt.show()
```
