# Import Image

```python
import cv2
img = cv2.imread("Mini.jpg") #Blue, Green, Red
```

```python
img.shape 
```

# Convert Image Colour Channel

```python
img = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
img.shape
```

# Resize the Image to 512x512

```python
img = cv2.resize(img,(512,512))
```

# Plot the Image

```python
import matplotlib.pyplot as plt
plt.imshow(img,cmap='gray') # Red, Blue, Green
plt.show()
```
